```python
import re

def extrair_secoes(conteudo):
  secoes = {}
  padrao = re.compile(r'([A-ZÁÉÍÓÚÂÊÎÔÛÃÕÇ\s]+):', re.MULTILINE)
  matches = list(padrao.finditer(conteudo))
  for i, match in enumerate(matches):
    titulo = match.group(1).strip()
    inicio = match.end()
    fim = matches[i+1].start() if i+1 < len(matches) else len(conteudo)

    conteudo_secao = conteudo[inicio:fim].strip()
    secoes[titulo] = conteudo_secao

  return secoes

with open("/content/laudo_A.txt", "r", encoding="utf-8") as f:
    laudoA = f.read()

with open("/content/laudo_B.txt", "r", encoding="utf-8") as f:
    laudoB = f.read()

for i, conteudo in enumerate([laudoA, laudoB], start=1):
  print(f"\n{'='*40}")
  print(f"LAUDO {i}")
  print(f"{'='*40}")

  secoes = extrair_secoes(conteudo)

  for titulo, texto in secoes.items():
        print(f"\n>> {titulo}:")
        print(texto)
```

```python
import unicodedata

campos_pop = [
    "descricao_vestes",
    "intervalo_post_mortem",
    "descricao_lesoes",
    "trajetoria_projetil",
    "coleta_vestigios",
    "causa_morte",
    "conclusao_fundamentada"
]

keywords_pop = {
    "descricao_vestes": [
        "veste", "vestes", "camiseta",
        "calca", "blusa", "roupa",
        "roupas", "vestimenta"
    ],
    "intervalo_post_mortem": [
        "post mortem", "intervalo", "rigidez",
        "livor", "temperatura", "horas", "estimado"
    ],
    "descricao_lesoes": [
        "lesao", "lesoes", "orificio",
        "ferimento", "regiao", "bordas",
        "diametro", "entrada"
    ],
    "trajetoria_projetil": [
        "trajetoria", "trajeto", "antero",
        "posterior", "descendente", "saida", "projetil"
    ],
    "coleta_vestigios": [
        "vestigio", "vestigios", "coleta",
        "cadeia", "custodia", "balistica", "fragmento"
    ],
    "causa_morte": [
        "causa", "morte", "obito", "hemorragia",
        "choque", "decorrente", "ferimento"
    ],
    "conclusao_fundamentada": [
        "conclusao", "morte violenta", "morte natural",
        "homicidio", "compativel"
    ]
}

def normalizar(texto):

  texto = unicodedata.normalize("NFD",texto)
  texto = "".join(c for c in texto if unicodedata.category(c) != "Mn")
  return texto.lower()

def verificar_conformidade(texto_laudo, campos_pop):
  texto_normalizado = normalizar(texto_laudo)
  resultado = {}

  for campo in campos_pop:
    keywords = keywords_pop.get(campo,[])
    encontrado = False

    for kw in keywords:
      kw_normalizada = normalizar(kw)
      if kw_normalizada in texto_normalizado:
        encontrado = True
        break

    resultado[campo] = encontrado
  return resultado

with open("/content/laudo_A.txt", "r", encoding="utf-8") as f:
    laudoA = f.read()

with open("/content/laudo_B.txt", "r", encoding="utf-8") as f:
    laudoB = f.read()


for i, conteudo in enumerate([laudoA, laudoB], start=1):
  print(f"\n{'='*40}")
  print(f"RELATÓRIO DE CONFORMIDADE - LAUDO {i}")
  print(f"{'='*40}")

  conformidade = verificar_conformidade(conteudo, campos_pop)

  for campo, presente in conformidade.items():
    status = "Presente" if presente else "Ausente"
    print(f"{campo:30s} {status}")



```

```python
import pandas as pd
import json
import re

def extrair_numero_laudo(texto):
  texto_norm = normalizar(texto)
  padrao = re.search(r'(laudo|necropsia|exame|n[o°º.])\s*[:\-.]?\s*n?[o°º.]?\s*(\d+[\/\-]?\d*)', texto_norm)
  if padrao:
    return padrao.group(2)
  return "Não identificado"

def extrair_sexo(texto):
  texto_norm = normalizar(texto)
  if any(p in texto_norm for p in ["masculino", "homem", "vitima do sexo masculino"]):
    return "Masculino"
  elif any(p in texto_norm for p in ["feminino", "mulher", "vitima do sexo feminino"]):
    return "Feminino"
  return "Não identificado"

def contar_lesoes(texto):
  texto_norm = normalizar(texto)
  numeros = re.findall(r'\b(\d+)\s*(lesao|lesoes|orificio|ferimento|perfuracao)', texto_norm)
  if numeros:
    return sum(int(n[0]) for n in numeros)
    # Conta ocorrências individuais das palavras-chave
  contagem = len(re.findall(r'\b(lesao|orificio|ferimento)\b', texto_norm))
  return contagem if contagem > 0 else 0

def calcular_conformidade_pct(conformidade):
    total = len(conformidade)
    presentes = sum(1 for v in conformidade.values() if v)
    return round((presentes / total) * 100, 1)

with open("/content/laudo_A.txt", "r", encoding="utf-8") as f:
    laudoA = f.read()

with open("/content/laudo_B.txt", "r", encoding="utf-8") as f:
    laudoB = f.read()

registros = []

for i, (nome, conteudo) in enumerate({"Laudo A": laudoA, "Laudo B": laudoB}.items(), start=1):
    conformidade = verificar_conformidade(conteudo, campos_pop)

    registro = {
        "numero_laudo":             extrair_numero_laudo(conteudo),
        "sexo_vitima":              extrair_sexo(conteudo),
        "descricao_vestes":         conformidade["descricao_vestes"],
        "intervalo_post_mortem":    conformidade["intervalo_post_mortem"],
        "num_lesoes":               contar_lesoes(conteudo),
        "conformidade_pct":         calcular_conformidade_pct(conformidade)
    }
    registros.append(registro)

# DataFrame
df = pd.DataFrame(registros, index=["Laudo A", "Laudo B"])

# Exibe
print(df)

# Salva JSON
df.to_json("laudos.json", orient="records", indent=4, force_ascii=False)
print("\nArquivo 'laudos.json' salvo com sucesso!")
```
